```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset using pd.merge  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# Reuse student data dataframe
mathdataset = student_data_df
# WORKS WORKS WORKS WORKS WORKS WORKS WORKS 
# mathgrouping=mathdataset.groupby(['school_name','grade'])
# avgbygrade=mathgrouping.mean()
# avgbygrade
# WORKS WORKS WORKS WORKS WORKS WORKS WORKS
# student_data_df.loc[(student_data_df['reading_score']>=70
# pass_math_byschool=student_data_df.loc[(student_data_df['grade']='9th')]
# mathdataset.tail()

#Retrieve needed columns only
mathset=mathdataset[['school_name','grade','math_score']]
mathset

# Locate 9th grade math scores, group by school name, then apply average function
ninth=mathset.loc[(mathdataset['grade']=='9th')]
ninthgroup=ninth.groupby(['school_name'])
ninth_df=pd.DataFrame(ninthgroup.mean())
ninth_df

# Locate 10th grade math scores, group by school name, then apply average function
tenth=mathset.loc[(mathdataset['grade']=='10th')]
tenthgroup=tenth.groupby(['school_name'])
tenth_df=pd.DataFrame(tenthgroup.mean())
tenth_df

# Locate 11th grade math scores, group by school name, then apply average function
eleventh=mathset.loc[(mathdataset['grade']=='11th')]
eleventhgroup=eleventh.groupby(['school_name'])
eleventh_df=pd.DataFrame(eleventhgroup.mean())
eleventh_df

# Locate 12th grade math scores, group by school name, then apply average function
twelfth=mathset.loc[(mathdataset['grade']=='12th')]
twelfthgroup=twelfth.groupby(['school_name'])
twelfth_df=pd.DataFrame(twelfthgroup.mean())
twelfth_df

# Start joining all of the dataframes by grade and renaming the columns
firstjoin=pd.merge(ninth_df,tenth_df,on='school_name')
rename1=firstjoin.rename(columns={'math_score_x':'9th','math_score_y':'10th'})
rename1
secondjoin=pd.merge(rename1,eleventh_df,on='school_name')
secondjoin
thirdjoin=pd.merge(secondjoin,twelfth_df,on='school_name')
thirdjoin
rename2=thirdjoin.rename(columns={'math_score_x':'11th','math_score_y':'12th'})
rename2
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins in which to place values
size_bins = [0, 583, 629, 644, 675]

#Create labels for the bins
bin_labels=['<$584','$585-629','$630-644','$645-675']

# Slice the data and place it into bins
schoolbin=all_school_data
pd.cut(schoolbin["Per Student Budget"], size_bins, labels=bin_labels).head()
schoolbin['Spending Ranges (per Student)']=pd.cut(schoolbin['Per Student Budget'],size_bins,labels=bin_labels)
schoolbin.head()


# Create a GroupBy object based upon "Spending Ranges"
spend_group=schoolbin.groupby('Spending Ranges (per Student)')
spend_group.head()

# Ungroup Needed Values from grouped dataframe
# Source: https://technology.amis.nl/2019/10/11/convert-groupby-result-on-pandas-data-frame-into-a-data-frame-using-to_frame/
new_spend_math=spend_group['Avg Math'].mean().to_frame(name='Avg Math Score').reset_index()
new_spend_read=spend_group['Avg Reading'].mean().to_frame(name='Avg Reading Score').reset_index()
new_spend_passmath=spend_group['% Passing Math'].mean().to_frame(name='% Passing Math').reset_index()
new_spend_passread=spend_group['% Passing Reading'].mean().to_frame(name='% Passing Reading').reset_index()
new_spend_overall=spend_group['% Overall Passing'].mean().to_frame(name='% Overall Passing').reset_index()
new_spend_overall

# Merge ungrouped dataframes
merge=pd.merge(new_spend_math,new_spend_read,on='Spending Ranges (per Student)')
merge2=pd.merge(merge,new_spend_passmath,on='Spending Ranges (per Student)')
merge3=pd.merge(merge2,new_spend_passread,on='Spending Ranges (per Student)')
merge4=pd.merge(merge3,new_spend_overall,on='Spending Ranges (per Student)')
merge4

# Remove index from printing
# Source: https://stackoverflow.com/questions/24644656/how-to-print-pandas-dataframe-without-index
merge4.style.hide_index()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Create bins in which to place values
school_size_bins = [0, 999, 2000, 5000]

#Create labels for the bins
school_size_bin_labels=['Small(<1000)','Medium(1000-2000)','Large(2000-5000)']

# Slice the data and place it into bins
schoolszebin=all_school_data
pd.cut(schoolszebin["Total Students"], school_size_bins, labels=school_size_bin_labels).head()
schoolszebin['School Size']=pd.cut(schoolbin['Total Students'],school_size_bins,labels=school_size_bin_labels)
schoolszebin.head()

# Create a GroupBy object based upon "Spending Ranges"
sze_group=schoolszebin.groupby('School Size')
sze_group.head()

# Ungroup Needed Values from grouped dataframe
# Source: https://technology.amis.nl/2019/10/11/convert-groupby-result-on-pandas-data-frame-into-a-data-frame-using-to_frame/
sze_math=sze_group['Avg Math'].mean().to_frame(name='Avg Math Score').reset_index()
sze_read=sze_group['Avg Reading'].mean().to_frame(name='Avg Reading Score').reset_index()
sze_passmath=sze_group['% Passing Math'].mean().to_frame(name='% Passing Math').reset_index()
sze_passread=sze_group['% Passing Reading'].mean().to_frame(name='% Passing Reading').reset_index()
sze_overall=sze_group['% Overall Passing'].mean().to_frame(name='% Overall Passing').reset_index()
sze_overall

# Merge ungrouped dataframes
szemerge=pd.merge(sze_math,sze_read,on='School Size')
szemerge2=pd.merge(szemerge,sze_passmath,on='School Size')
szemerge3=pd.merge(szemerge2,sze_passread,on='School Size')
szemerge4=pd.merge(szemerge3,sze_overall,on='School Size')
szemerge4

# Remove index from printing
# Source: https://stackoverflow.com/questions/24644656/how-to-print-pandas-dataframe-without-index
szemerge4.style.hide_index()
```
